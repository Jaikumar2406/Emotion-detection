```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    horizontal_flip=True,
    zoom_range=0.2
)
```

```python
valiodation_datagen = ImageDataGenerator(rescale=1./255)
```

```python
data_dir = 'images/'
train = 'images/train/'
validation = 'images/validation/'
```

```python
train_genrator = train_datagen.flow_from_directory(
    train,
    target_size=(128,128),
    batch_size=16 , 
    class_mode='categorical',
    color_mode='grayscale')
```

```python
validation_genrator = valiodation_datagen.flow_from_directory(
    validation,
    target_size=(128,128),
    batch_size=16,
    color_mode='grayscale',
    class_mode='categorical')
```

```python
class_names = list(train_genrator.class_indices.keys())
```

```python
class_names
```

```python
images, labels = next(train_genrator)
plt.figure(figsize=(10, 10))
for i in range(9):
    plt.subplot(3, 3, i + 1)
    plt.imshow(images[i].reshape(128, 128), cmap='gray')
    plt.title(class_names[np.argmax(labels[i])])
    plt.axis('off')
plt.tight_layout()
plt.show()
```

## cnn model

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense , MaxPooling2D , Conv2D , Flatten , BatchNormalization , Dropout
from tensorflow.keras.models import Sequential
```

```python
model = Sequential()
model.add(Conv2D(64 , (3, 3) , activation='relu' , input_shape = (128,128,1))),
model.add(MaxPooling2D(pool_size=(3 , 3))),
model.add(BatchNormalization()),
model.add(Dropout(0.2))

model.add(Conv2D(128 , (3,3) , activation='relu')),
model.add(MaxPooling2D(pool_size=(3,3))),
model.add(BatchNormalization()),
model.add(Dropout(0.2)),


model.add(Conv2D(256 , (3,3) , activation='relu')),
model.add(MaxPooling2D(pool_size=(3,3))),
model.add(BatchNormalization()),
model.add(Dropout(0.2))

model.add(Flatten()),
model.add(Dense(256 , activation='relu')),
model.add(Dropout(0.5)),
model.add(BatchNormalization()),

model.add(Dense(7 , activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint,ReduceLROnPlateau
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=1e-6, verbose=1)

# Model Training
history = model.fit(
    train_genrator,
    epochs=20,
    validation_data=validation_genrator,
    callbacks=[early_stopping, lr_scheduler]
)
```

```python
model.save('model1.h5')
print("saved!!!")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image

def predictions(img_path, model, class_names):
    # Load and preprocess image
    img = image.load_img(img_path, target_size=(128, 128), color_mode='grayscale')
    img_array = image.img_to_array(img) / 255.0  
    img_array = np.expand_dims(img_array, axis=0) 

    # Predict the class
    predicted = model.predict(img_array)
    prediction_class = np.argmax(predicted)
    confidence = round(predicted[0][prediction_class] * 100, 2)

    # Display image
    plt.imshow(img, cmap='gray')
    plt.axis('off')
    plt.show()

    return class_names[prediction_class], confidence 
```

```python
prediction_label, confidence = predictions("601.jpg", model, class_names)
print(prediction_label, confidence)
```
